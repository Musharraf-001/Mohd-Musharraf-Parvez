```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
import plotly.colors as colors
pio.templates.default = "plotly_white"
```

```python
data = pd.read_csv("Sample - Superstore.csv", encoding = 'latin-1')
data.head()
```

```python
data.describe()
```

```python
data.info()
```

Converting date format


```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
data['Ship Date'] = pd.to_datetime(data['Ship Date'])
```

```python
data.head()
```

```python
data['Order Month'] = data['Order Date'].dt.month
data['Order Year'] = data['Order Date'].dt.year
data['Order Day of Week'] = data['Order Date'].dt.dayofweek
```

```python
data.head()
```

# Monthly Sales Analysis

```python
sales_by_month = data.groupby('Order Month')['Sales'].sum().reset_index()
fig = px.line(sales_by_month, x='Order Month', y='Sales', title='Monthly Sales Analysis')
fig.show
```

```python
sales_by_month = data.groupby('Order Month')['Sales'].sum().reset_index()
print(sales_by_month)
```

## Sales by category

```python
sales_by_category = data.groupby('Category')['Sales'].sum().reset_index()
fig = px.pie(sales_by_category, values='Sales', names='Category', hole=0.2, title='Sales by Category')
fig.show()
```

# Data analysis by sub category

```python
sales_by_sub_category = data.groupby('Sub-Category')['Sales'].sum().reset_index()
fig = px.bar(sales_by_sub_category, x='Sub-Category', y='Sales', title='Sales by Sub-Category')
fig.show()
```

# Monthly Profit Analysis




```python
profit_by_month = data.groupby('Order Month')['Profit'].sum().reset_index()
profit_by_month
```

```python
fig = px.line(profit_by_month, x = 'Order Month', y = 'Profit', title = 'Monthly Profit Analysis')
fig.show()
```

```python
fig = px.bar(profit_by_month, x = 'Order Month', y = 'Profit', title = 'Monthly Profit Analysis')
fig.show()
```

# Profit By Category

```python
profit_by_category = data.groupby('Category')['Profit'].sum().reset_index()
profit_by_category
```

```python
fig = px.pie(profit_by_category,
             values='Profit',
             names='Category',
             hole=0.5,
             color_discrete_sequence=px.colors.qualitative.Pastel)

fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(title='Profit by Category', title_font=dict(size=24))
fig.show()
```

# Profit by sub category

```python
profit_by_subcategory = data.groupby('Sub-Category')['Profit'].sum().reset_index()
profit_by_subcategory
```

```python
fig = px.bar(profit_by_subcategory, x = 'Sub-Category', y = 'Profit', title = 'Profit by Sub-Category')
fig.show()
```
